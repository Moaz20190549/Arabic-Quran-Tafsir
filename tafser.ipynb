```python
def remove_diacritics(text):
    arabic_diacritics = re.compile("""
                                ّ    | # Tashdid
                                َ    | # Fatha
                                ً    | # Tanwin Fath
                                ُ    | # Damma
                                ٌ    | # Tanwin Damm
                                ِ    | # Kasra
                                ٍ    | # Tanwin Kasr
                                ْ    | # Sukun
                                ـ     # Tatwil/Kashida
                            """, re.VERBOSE)
    return re.sub(arabic_diacritics, '', text)
```

```python
def normalize_arabic(text):
    text = re.sub("[إأآا]", "ا", text)
    text = re.sub("ى", "ي", text)
    text = re.sub("ؤ", "و", text)
    text = re.sub("ئ", "ي", text)
    text = re.sub("ة", "ه", text)
    return text
```

```python
def remove_stopwords(text):
    words = text.split()
    clean_words = [word for word in words if word not in arabic_stopwords]
    return ' '.join(clean_words)
```

```python
def preprocess_tafsir_data(df):
    df.dropna(subset=['ayah', 'tafsir_content'], inplace=True)

    df.drop_duplicates(inplace=True)

    df['ayah'] = df['ayah'].apply(lambda x: remove_diacritics(x))
    df['ayah'] = df['ayah'].apply(lambda x: normalize_arabic(x))
    
    df['tafsir_content'] = df['tafsir_content'].apply(lambda x: remove_diacritics(x))
    df['tafsir_content'] = df['tafsir_content'].apply(lambda x: normalize_arabic(x))

    df['ayah'] = df['ayah'].str.strip()
    df['tafsir_content'] = df['tafsir_content'].str.strip()
    
    return df
```

```python
def preprocess_text(text):
    text = remove_diacritics(text)
    text = re.sub(r'[^\w\s]', '', text)  
    text = text.strip().lower() 
    return text
```

```python
def find_closest_ayah(user_query, df):
    df = preprocess_tafsir_data(df)
    
    preprocessed_query = preprocess_text(user_query)  
    
    df['processed_ayah'] = df['ayah'].apply(preprocess_text)  
    
    tfidf = TfidfVectorizer()
    ayah_vectors = tfidf.fit_transform(df['processed_ayah'])
    
    query_vector = tfidf.transform([preprocessed_query])
    
    similarities = cosine_similarity(query_vector, ayah_vectors)
    
    closest_idx = similarities.argmax()
    
    closest_ayah_text = df.iloc[closest_idx]['ayah']
    
    matching_ayahs = df[df['ayah'] == closest_ayah_text]
    
    result_df = matching_ayahs[['surah_name','ayah', 'tafsir_content', 'tafsir_book']]
    
    return result_df
```
